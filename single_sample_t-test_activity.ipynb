#### Requirements
Determine whether a miles per gallon (mpg) rating of 40 is unusual for a hybrid car on the market in 2013. 

#### Import packages

```python
import pandas as pd
from scipy import stats
from scipy.stats import norm
```

#### Import data

```python
hybrid2013 = pd.read_csv('C:/Users/desja/python_course/datasets/hybrid2013.csv')
hybrid2013
```

#### Test Assumptions

```python
hybrid2013['mpg'].hist()
```

#### Results
Looks like things are only slightly normally distributed

#### Run the Analysis

```python
stats.ttest_1samp(hybrid2013['mpg'], 40)
```

#### Results
The p value is significantly lower at p < .05. That means that the mpg rating of 40 is different than the mean for hybrid cars. To check if that rating is higher or lower I will need to examine the population mean:

```python
hybrid2013.mpg.mean()
```

#### Results
It turns out that the average mpg rating for a hyrbid car in 2013 was 33 so driving a car with a rating of 40 would have been a very good deal.
