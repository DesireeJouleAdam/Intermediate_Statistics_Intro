#### Requirements
Determine if the mean miles per gallon for a compact and a large car differ from each other.

#### Import Packages

```python
import pandas as pd
import numpy as np
from scipy import stats
from scipy.stats import ttest_ind
```

#### Import Data

```python
hybrid2013 = pd.read_csv('C:/Users/desja/python_course/datasets/hybrid2013.csv')
hybrid2013
```

#### Test Assumptions

```python
hybrid2013.mpg[hybrid2013.carclass == 'C'].hist()
```

```python
hybrid2013.mpg[hybrid2013.carclass == 'L'].hist()
```

#### Results
Distribution is not normal

#### Run the Analysis

```python
ttest_ind(hybrid2013.mpg[hybrid2013.carclass == 'C'], hybrid2013.mpg[hybrid2013.carclass == 'L'])
```

#### Results 

Looks like there is a significant difference between compact and large hybrid cars in terms of miles per gallon, since the p value is less than .05.

I can examine the mean for each group to see the difference in the means.

```python
hybrid2013.mpg[hybrid2013.carclass == 'C'].mean()
```

```python
hybrid2013.mpg[hybrid2013.carclass == 'L'].mean()
```

#### Results
The mean miles per gallon for a compact and a large car differ from each other with a large car achieving less miles per gallon. You would get more miles per gallon in a compact car in 2013.
